```python
## Networks and Gels, Serie 2 - Exercise 3b
import warnings 
warnings.filterwarnings('ignore')
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True
%matplotlib inline
```

```python
def xSpinoidal(n1,n2,phi):
    return 0.5 /(n1*phi) + 0.5 / (n2*(1-phi))

def p2array(n1,n2):
    y = []
    x = []
    for i in np.arange(0.00,1,0.01):
        x.append(i)
        y.append(xSpinoidal(n1,n2,i))
    return x,y

def graph(idx,n1,n2):
    plt.title("Plot %s (n1 = %s, n2 = %s)"%(idx,n1,n2))
    x,y = p2array(n1,n2)
    plt.plot(x,y,color='b')
    plt.xlabel('$\phi$')
    plt.ylabel('$\chi$')
    plt.savefig('3b_%s.png'%(idx))
    plt.show()

def allPlots():
    graph('i',1,1)
    graph('ii',1,10000)
    graph('iii',10000,10000)
    graph('iv',100,10000)
```

```python
allPlots()
```
